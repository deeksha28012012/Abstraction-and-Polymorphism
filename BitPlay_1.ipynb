```python
def numberofBits(n):
  ones = 0
  zeroes = 0
  while(n):
    if(n & 1 == 1):
      ones+= 1
    else:
      zeroes+= 1
    n>>=1
  print("number of ones: ",ones,"number of zeroes: ",zeroes)
number= int(input("enter a number:"))
numberofBits(number)
```

```python
def setornot(number, n):
  mask = 1
  if (n & mask)==1 or (n & mask) == 0:
    if number & (1<<(n-1)):
      print("SET")
    else:
      print("NOT SET")
number=int(input("enter a number: "))
n = int(input("enter the bit position: "))
setornot(number, n)
```
