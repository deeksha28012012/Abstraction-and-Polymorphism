```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('gapminder(2007) (2).csv')
```

```python
data.head()
```

```python
data.info()
```

```python
avg_data = data.groupby(data['continent']).mean(numeric_only=True)
```

```python
avg_data = avg_data.reset_index()
```

```python
avg_data
```

```python
plt.bar(avg_data['continent'], avg_data['life_exp'], color='teal')
plt.xlabel('Continent')
plt.ylabel('Average Life Expectancy')
plt.show()
```

```python
plt.bar(avg_data['continent'], avg_data['gdp_cap'], color='darkred')
plt.xlabel('continent')
plt.ylabel('Average GDP')
plt.show()
```

```python
sns.countplot(x=data['continent'], palette='autumn')
```
