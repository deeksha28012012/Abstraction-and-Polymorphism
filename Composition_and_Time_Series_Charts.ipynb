```python
data.head()
```

```python
data.groupby('continent').size().plot(kind='pie',autopct='%.2f')
```

```python
plt.pie(data.groupby('continent').size(), autopct='%.2f', labels = ['Africa', 'America', 'Asia', 'Europe','Oceania'], labeldistance=1.15, wedgeprops= {'linewidth':2, 'edgecolor':'white'})
```
