```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Titanic Dataset (4).csv')
```

```python
data.head(5)
```

```python
sns.countplot(x='Gender', hue='Survived', data=data)
```

```python
sns.countplot(x='Pclass', hue='Survived', data=data)
```

```python
sns.distplot(data['Age'], kde=False, bins=40)
```

```python
sns.countplot(data['Gender'])
```

```python
sns.countplot(x='Survived', hue='SibSp', data=data, palette='mako')
```

```python
sns.countplot(x='Survived', hue='Parch', data=data, palette='mako')
```

```python
sns.distplot(data['Fare'])
plt.show()
```

```python
sns.boxplot(x='Pclass', y='Age', data=data, palette='spring')
```

```python
numerical_data = data.select_dtypes(include=['number'])
sns.heatmap(numerical_data.corr(), annot=True, cmap='autumn')
plt.show()
```
