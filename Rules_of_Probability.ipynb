```python
set1 = {'A', 'B', 'C', 'D', 'E'}
set2 = {'B', 'D', 'V', 'X', 'Y', 'Z'}
union = set1.union(set2)
total_guests = list(union)
print("total guests to be invited", len(total_guests))
print('guest list:', total_guests)
```

```python
set1 = {'A', 'B', 'C', 'D', 'E'}
set2 = {'B', 'D', 'V', 'X', 'Y', 'Z'}
intersection = set1.intersection(set2)
total_guests= list(intersection)
print("total guests to be invited", len(total_guests))
print('guest list:', total_guests)
```

```python
def prob_a_or_prob_b(a, b, all_possible_outcomes):
  prob_a = len(a)/len(all_possible_outcomes)
  prob_b = len(b)/len(all_possible_outcomes)
  inter = a.intersection(b)
  prob_inter = len(inter)/len(all_possible_outcomes)
  return(prob_a + prob_b - prob_inter)
evens = {2,4,6}
greater_than_two = {3,4,5,6}
all_possible_rolls = {1,2,3,4,5,6}
print('possibilty of getting an even number or a number greater than two: ')
print(prob_a_or_prob_b(evens, greater_than_two, all_possible_rolls))
```
