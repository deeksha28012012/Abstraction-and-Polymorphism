```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats
```

```python
data = pd.read_csv('Titanic Dataset.csv')
data.head(5)
```

```python
data.dtypes
```

```python
mean_age = np.mean(data['Age'])
print("Mean Age is = ", mean_age)
```

```python
mean_fare = np.mean(data['Fare'])
print("Mean Fare is = ", mean_fare)
```

```python
median_age = np.median(data['Age'])
print("Meadian Age is = ", median_age)
```

```python
median_fare = np.median(data['Fare'])
print("Meadian Fare is = ", median_fare)
```

```python
mode_age = stats.mode(data['Age'])
print("Mode Age is = ", mode_age)
```

```python
mode_class = stats.mode(data['Pclass'])
print("Mode Class is = ", mode_class)
```
