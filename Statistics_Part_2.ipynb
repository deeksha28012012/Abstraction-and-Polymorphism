```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as stats
```

```python
data = pd.read_csv('Weather Dataset.csv')
```

```python
data.head(5)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
mean_temp = np.mean(data['Temperature (C)'])
print(mean_temp)
```

```python
median_temp = np.median(data['Temperature (C)'])
print(median_temp)
```

```python
var_temp = np.var(data['Temperature (C)'])
print(var_temp)
```

```python
standard_deviation_temp = np.std(data['Temperature (C)'])
print(standard_deviation_temp)
```
