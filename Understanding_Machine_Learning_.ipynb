```python
import nltk
from nltk.chat.util import reflections, Chat
```

```python
reflections = {
  "i am"     : "you are",
  "i was"    : "you were",
  "i"        : "you",
  "i'm"      : "you are",
  "i'd"      : "you would",
  "i've"     : "you have",
  "i'll"     : "you will",
  "my"       : "your",
  "you are"  : "I am",
  "you were" : "I was",
  "you've"   : "I have",
  "you'll"   : "I will",
  "your"     : "my",
  "yours"    : "mine",
  "you"      : "me",
  "me"       : "you"
}

pairs = [
    [
        r"my name is (.*)",
        ["Hello %1, How are you today?",]
    ],
    [
        r"hi|hello|hey",
        ["hello", "hey there",]
    ],
    [
        r"What's your name?",
        ["I am a bot created by Deeksha, you can call me Jarvis though!",]
    ],
    [
        r"how are you?",
        ["I am doing great and how about you, mate?",]
    ],
    [
        r"sorry (.*)",
        ["Its alright", "Its okay, Nevermind",]
    ]
]
```

```python
def chat():
  print("Hi! I am a chatbot created by deeksha for your service")
  chat = Chat(pairs, reflections)
  chat.converse()
if __name__ == "__main__":
  chat()
```
