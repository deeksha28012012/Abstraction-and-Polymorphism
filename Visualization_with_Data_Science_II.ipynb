```python
import matplotlib.pyplot as plt
x=[0, 5, 10, 15, 20, 25, 30]
y1=[10, 15, 20, 20, 30, 15, 0]
y2=[10, 12, 15, 12, 20, 10, 0]
plt.plot(x,y1, linestyle='dashed', marker='o')
plt.plot(x,y2, linestyle='dashed', marker='o')
plt.title('VELOCITY TIME GRAPH')
plt.xlabel('Velocity m/s')
plt.ylabel('Time (s)')
plt.xlim(5, 25)
plt.ylim(5, 25)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,11,1)
y1=(2*x)+1
y2=(2*x*x)+2
plt.plot(x, y1, 'indigo', linewidth=20, label ='y*2+1')
plt.plot(x, y2, 'pink', linewidth=20, label ='2x^2+1')
plt.legend()
plt.show()
```
